```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import scipy.linalg as la
import math
```

```python
ref = pd.read_excel("ref.xlsx")
font = {'family' : 'Arial',
        'weight' : 'normal',
        'size'   : 18}
matplotlib.rc('font', **font)

plt.figure(figsize=(12,6))
plt.plot(ref.Frequency/1e6, ref.S21, linewidth=2)
plt.xlabel('Frequency (MHz)')
plt.ylabel('S21 (dB)')
plt.grid(True, which="both", ls="-")
plt.title('Background/Noise - just two unconnected cable far away from each other')
```

```python
font = {'family' : 'Arial',
        'weight' : 'normal',
        'size'   : 18}
matplotlib.rc('font', **font)

ch1 = pd.read_excel("ch1.xlsx")
ch1_shielded = pd.read_excel("channel1shielded.xlsx")

plt.figure(figsize=(12,6))
plt.plot(ch1.Frequency/1e6, ch1.S21, linewidth=2)
plt.plot(ch1_shielded.Frequency/1e6, ch1_shielded.S21, linewidth=2)

plt.xlabel('Frequency (MHz)')
plt.ylabel('S21 (dB)')
plt.grid(True, which="both", ls="-")
plt.legend(['Dsub unshielded', 'Dsub shielded'], loc='upper left')
plt.title('S21 through channel 1')
```

```python
font = {'family' : 'Arial',
        'weight' : 'normal',
        'size'   : 18}
matplotlib.rc('font', **font)

ch1 = pd.read_excel("crosstalk.xlsx")
ch1_shielded = pd.read_excel("crosstalk_shielded.xlsx")

plt.figure(figsize=(12,6))
plt.plot(ch1.Frequency/1e6, ch1.S21, linewidth=2)
plt.plot(ch1_shielded.Frequency/1e6, ch1_shielded.S21, linewidth=2)

plt.xlabel('Frequency (MHz)')
plt.ylabel('S21 (dB)')
plt.grid(True, which="both", ls="-")
plt.legend(['Dsub unshielded', 'Dsub shielded'], loc='upper left')
plt.title('Crosstalk between ch1 and ch3: Input-Ch1, Output-Ch3')
```
