```python
import nidaqmx
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
from nidaqmx.constants import AcquisitionType as AcqType
from nidaqmx.constants import TerminalConfiguration as TermConfig

rs_acqu = 1e6 # sampling rate
t_acqu = 5 # acquisition time
ns_acqu = int(rs_acqu*t_acqu)
t = np.arange(start=0, stop=5, step=1/rs_acqu)

# Femto Preamp Channel
with nidaqmx.Task() as acqu_task:
    # Set acquisition channels and specs
    acqu_task.ai_channels.add_ai_voltage_chan("Dev1/ai0",terminal_config=TermConfig.DIFFERENTIAL,min_val=-1,max_val=1)
    acqu_task.timing.cfg_samp_clk_timing(rs_acqu,sample_mode=AcqType.FINITE,samps_per_chan=ns_acqu)

    # Start acquisition
    acqu_task.start()
    
    # Collect the data measured
    i0 = acqu_task.read(number_of_samples_per_channel=ns_acqu)
    
    # Stop acquisition
    acqu_task.stop()

    # Plot data 
titlestr='No ground'
#v0f=np.array(v0)
#v1f=np.array(v1)
i0f=np.array(i0)/1e8

# Femto Preamp Channel 
plt.figure()
plt.plot(t, i0f)
plt.xlabel('t [s]')
plt.ylabel('i [A]')
plt.title(titlestr)
plt.ylim()

f,II_den=signal.periodogram(i0f,fs=rs_acqu,window='flattop',scaling='density',return_onesided=True)
plt.figure()
plt.loglog(f,np.sqrt(II_den))
plt.xlabel('f[Hz]')
plt.ylabel('S_{ii}(f)^{1/2} [A.Hz^{-1/2}]')
plt.title(titlestr)
```
